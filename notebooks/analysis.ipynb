```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import coffee data
coffee = pd.read_csv("../data/crop/coffee_output.csv")


# import weather data
weather = pd.read_csv('../data/weather/weather_data.csv')
```

```python
# TODO: Merge these two dataframes together

#using inner join merge so only join when they have in common which is year and coffee
merge_df = pd.merge(weather, coffee, on = ['year','subdivision'])
merge_df
```

```python
print(f"Shape of Merged Data Set {merge_df.shape}")
#print(f"Shape of Merged Data Set {coffee_weather_merged.shape}")
```

```python
# TODO: View the first 5 rows of your data

merge_df.head()
```

```python
# TODO: Check for null values 
merge_df.isna().sum()
```

```python
# TODO: Drop rows that contain null values

merge_df = merge_df.dropna()
merge_df
```

```python

#Relationship between temp_avg and 60kgs_bag

#Pivot table
print("\nPivot table between Average Temperature and 60 kgs bags:")
pivot = pd.pivot_table(merge_df, values=['temp_avg','60kgs_bag'], index = 'year', aggfunc= 'mean')
display(pivot)

#Scatterplot 
sns.scatterplot(merge_df, x = "temp_avg", y  = "60kgs_bag", color = "blue")
plt.title('Relationship between Average Temperature and 60 kgs bag in Minas Gerais ')
plt.xlabel('Average Temperature')
plt.ylabel('60 kg bags of coffee beans harvested\n (million bags)')
plt.show()

#Correlation
tempavg_60kgs = merge_df[["temp_avg", "60kgs_bag"]].corr()
print(f"Correlation between temp_avg and 60kgs_bag: \n {tempavg_60kgs}")

#Correlation matrix
sns.heatmap(tempavg_60kgs, annot=True, cmap='viridis').set_title('Correlation')
plt.tight_layout()
plt.show()
```

```python
#Relationship between minimum humidity and 60kgs_bag

#scatterplot which shows relationship between minimum humidity and 60kgs_bag
sns.scatterplot(merge_df, x = "hum_min", y  = "60kgs_bag", color = "green")
plt.title('Relationship between Minimum Humidity and 60kgs_bag')
plt.xlabel('Minimum Humidity')
plt.ylabel('60 kg bags of coffee beans harvested (million bags)')
plt.show()

#Correlation
minhum_60kgs = merge_df[["hum_min", "60kgs_bag"]].corr()
print(f"Correlation between Minimum Humidity and 60kgs_bag: \n {minhum_60kgs}")

#Correlation Matrix
sns.heatmap(minhum_60kgs, annot=True, cmap='viridis').set_title('Correlation')
plt.tight_layout()
plt.show()
```

```python
#Relationship between rain_max and 60kgs_bag

#scatterplot of the relationship between rain_max and 60kgs_bag
sns.scatterplot(merge_df, x = "rain_max", y  = "60kgs_bag", color = "orange")
plt.title('Relationship between Maximum Rain and 60kgs_bag')
plt.xlabel('Maximum Rain')
plt.ylabel('60 kg bags of coffee beans harvested (million bags)')
plt.show()

#correlation
rainmax_60kgs = merge_df[["rain_max", "60kgs_bag"]].corr()
print(f"Correlation between Maximum Rain and 60kgs_bag: \n {rainmax_60kgs}")

# correlation matrix
sns.heatmap(rainmax_60kgs, annot=True, cmap='viridis').set_title('Correlation')
plt.tight_layout()
plt.show()
```

```python
#Relationship between average wind speed and 60kgs_bag

#scatterplot between average wind speed and 60kgs_bag
sns.scatterplot(merge_df, x = "wind_avg", y  = "60kgs_bag", color = "purple")
plt.title('Relationship between average wind speed and 60kgs_bag')
plt.xlabel('average wind speed')
plt.ylabel('60 kg bags of coffee beans harvested (million bags)')
plt.grid(True)
plt.show()

#correlation
windspeed_60kgs = merge_df[["wind_avg", "60kgs_bag"]].corr()
print(f"Correlation between average wind speed and 60kgs_bag: \n {windspeed_60kgs}")

#Correlation matrix
sns.heatmap(windspeed_60kgs, annot=True, cmap='viridis').set_title('Correlation')
plt.tight_layout()
plt.show()
```

```python
#Bivariate analysis between the mentioned weather columns (temp_avg, hum_min, rain_max, and wind_avg) and bearing_trees.

#bearing_trees correlation
bt_corr_merge = merge_df[["rain_max","hum_min", "temp_avg","hum_max","wind_avg", "bearing_trees"]].corr()
print(f"Correlation between Weather demographics and bearing_trees: \n {bt_corr_merge}")

#bearing_trees correlation matrix
sns.heatmap(bt_corr_merge, annot=True, cmap='viridis').set_title('Correlation')
plt.tight_layout()
plt.show()
```

```python
# Correlation between Weather demographics and 60kgs_bag

#Correlation
corr_merge = merge_df[["rain_max","hum_min", "temp_avg","hum_max","wind_avg", "60kgs_bag"]].corr()
print(f"Correlation between Weather demographics and 60kgs_bag: \n {corr_merge}")

#Correlation matrix
sns.heatmap(corr_merge, annot=True, cmap='viridis').set_title('Correlation')
plt.tight_layout()
plt.show()
```
