**Univariate Analysis**

This is a univariate analysis across the years for Minas Gerais of the following columns: 
* bearing trees
* bearing_hectares
* 60kgs_bag


```python
#bearing_trees 
#bearing_trees: Amount of bearing coffee trees (million trees)

#This makes the year column as an actual date 
coffee_mg["year"] = pd.to_datetime(coffee_mg["year"], format='%Y') 

#line graph to show how bearing trees are changing over time 
coffee_mg.plot.line(x= "year", y = "bearing_trees") 
plt.xlabel('Year')
plt.ylabel('Bearing Trees (million trees)')
plt.title('Rate of Bearing Trees over Years in Minas Gerais ')
plt.grid(True)
plt.show()


#histogram to show distributions of bearing trees
#coffee_mg["bearing_trees"].plot.hist(bins= 30)
sns.histplot(data=coffee_mg, x="bearing_trees", kde=True, color= "green")
plt.xlabel('Bearing Trees (million trees)')
plt.title(' Distribution of Bearing Trees in Minas Gerais')
plt.show() 

coffee_mg[["bearing_trees"]].agg(['mean', 'median'])
```

```python
#Bearing Hectares
#bearing hectares = Hectares of bearing coffee trees (thousand hectares)

#line graph to to show how hectares are changing over time 
coffee_mg.plot.line(x= "year", y = "bearing_hectares", color = "purple") 
plt.xlabel('Year')
plt.ylabel('Bearing Hectares (thousand hectares)')
plt.title('Rate of Bearing Hectares over Years in Minas Gerais')
plt.grid(True)
plt.show()

#histogram to show distributions of bearing hectares
sns.histplot(data=coffee_mg, x="bearing_hectares", kde=True, color= "green", bins = 15)
plt.xlabel('Bearing Hectares (thousand hectares)')
plt.title('Distribution of Bearing Hectares in Minas Gerais')
plt.show() 

#aggregate functions of bearing hectares
coffee_mg[["bearing_hectares"]].agg(['mean', 'median'])


```

```python
#60kgs bags of coffee 
#60kgs_bag: 60 kg bags of coffee beans harvested (million bags)

#line graph to to show how 60 kg bags of coffee beans are changing over time 
coffee_mg.plot.line(x= "year", y = "60kgs_bag") 
plt.xlabel('Year')
plt.ylabel('60 kg bags of coffee beans harvested\n (million bags)')
plt.title('Rate of 60 kg bags of coffee beans harvested\n over Years in Minas Gerais')
plt.grid(True)
plt.show()


#histogram to show distributions of 60 kg bags of coffee beans
sns.histplot(data=coffee_mg, x="60kgs_bag", kde=True, color= "blue", bins = 20)
plt.xlabel('60 kg bags of coffee beans harvested (million bags)')
plt.title('Distribution of 60 kg bags of coffee beans\n harvested in Minas Gerais')
plt.show() 

#aggregate functions of 60 kg bags of coffee beans
coffee_mg[["60kgs_bag"]].agg(['mean', 'median'])

```

```python
# What is the relationship between bearing_trees and 60kgs_bag? 
#How *strong* is this relationship? Use pearson's correlation coefficient.

print("\nPivot table between bearing_trees and 60kgs_bag:")
pivot = pd.pivot_table(coffee_mg, values=['bearing_trees','60kgs_bag'], index = 'year')
display(pivot)

sns.scatterplot(coffee_mg, x = "bearing_trees", y  = "60kgs_bag", color = "blue")
plt.title('Relationship between bearing_trees and 60kgs_bag in in Minas Gerais')
plt.xlabel('Bearing Trees (million trees)')
plt.ylabel('60 kg bags of coffee beans harvested (million bags)')
plt.show()

trees_60kgs = coffee_mg[["bearing_trees", "60kgs_bag"]].corr()
print(f"Correlation between bearing_trees and 60kgs_bag: \n {trees_60kgs}")

sns.heatmap(trees_60kgs, annot=True, cmap='viridis').set_title('Correlation')
plt.tight_layout()
plt.show()

     

```

```python
#Relationship between bearing_hectares and 60kgs_bag? 

#Pivot table 
print("\nPivot table between bearing_hectares and 60kgs_bag:")
pivot = pd.pivot_table(coffee_mg, values=['bearing_hectares','60kgs_bag'], index = 'year', aggfunc= 'mean')
display(pivot)

#Scatterplot 
sns.scatterplot(coffee_mg, x = "bearing_hectares", y  = "60kgs_bag", color = "blue")
plt.title('Relationship between bearing_hectares and 60kgs_bag in Minas Gerais ')
plt.xlabel('bearing_hectares')
plt.ylabel('60 kg bags of coffee beans harvested\n (million bags)')
plt.show()

#Correlation
hectares_60kgs = coffee_mg[["bearing_hectares", "60kgs_bag"]].corr()
print(f"Correlation between bearing_hectares and 60kgs_bag: \n {hectares_60kgs}")

#Correlation matrix
sns.heatmap(hectares_60kgs, annot=True, cmap='viridis').set_title('Correlation')
plt.tight_layout()
plt.show()
```
