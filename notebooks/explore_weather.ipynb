```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")

# Check the shape of each data file 
print(f"Shape of Weather 1 {weather1.shape}")
print(f"Shape of Weather 2 {weather2.shape}")

#merged data file
weather = pd.concat([weather1,weather2],ignore_index=True)
print(f"Shape of Weather  {weather.shape}")


weather.to_csv('../data/weather/weather_data.csv', index=False)
weather
```

**Univariate Analysis**

This is a univariate analysis across the years for Minas Gerais of the following columns: 
* average temperature
* minimum humidity
* average wind speed 

This will also look at the highest maximum record rainfall. This analysis will include line graphs. 

```python
#Average Temperature

#This makes the year column as an actual date 
weather["year"] = pd.to_datetime(weather["year"], format='%Y') 

#This line graph is showing how the average temperature (Celsius) is changing across the years at Minas Gerais. 
weather.plot.line(x= "year", y = "temp_avg", color='red') 
plt.xlabel('Year')
plt.ylabel('Average temperature (C)')
plt.title('Average Temperature of (C) Minas Gerais over Years')
plt.grid(True)
plt.show()

#Max and Min Average Temperature
max_avg_temp = weather[["temp_avg"]].sort_values(by = "temp_avg", ascending = False).head()
print(f"Top 5 Years with Maximum Average Temperature: \n {max_avg_temp}")

min_avg_temp = weather[["temp_avg"]].sort_values(by = "temp_avg", ascending = True).head()
print(f"\nTop 5 Years with Minimum Average Temperature: \n {min_avg_temp}")

```

```python
#Minimum Humidity 

# This line graph is showing the minimum humidity across the years in Minas Gerais
min_humidity_year = weather.plot.line(x= "year", y = "hum_min", color='orange')
plt.xlabel('Year')
plt.ylabel('Minimum Humidity %')
plt.title('Minimum Humidity of Minas Gerais over Years')
plt.grid(True)
plt.show()

```

```python
# This line graph is showing the average wind speed across the years

weather.plot.line(x= "year", y = "wind_avg") 
plt.xlabel('Year')
plt.ylabel('Average Wind Speed (meters per second)')
plt.title('Average Wind Speed of Minas Gerais over Years')
plt.grid(True)
plt.show()

#This is calculating the correlation between average wind speed and the year
avg_wind_corr = weather[["year", "wind_avg"]].corr()
print(f"Correlation between Average Wind Speed and Year: \n {avg_wind_corr}")
```

```python
#highest maximum record rainfall

year_max_rain = weather[["rain_max"]].sort_values(by= "rain_max", ascending = False).head()
year_max_rain

#getting the first index - The year 2006 had the highest maximum record rainfall 
weather.iloc[[0]]

```

**Bivariate Analysis**

This will be analyzing the relationship between the following dimensions:

* max rain and max humidity
* average wind speed and average temperature

This analysis will be done utilizing pivot tables, scatter plots, and Pearson's coefficient. 

```python
#Relationship between max rain and max humidity

#pivot table 
print("\nPivot table between max rain and max humidity:")
pivot = pd.pivot_table(weather, values=['rain_max','hum_max'], index = 'year', aggfunc= 'mean')
display(pivot)

#scatterplot
sns.scatterplot(weather, x = "rain_max", y  = "hum_max", color = "purple")
plt.title('Relationship between Max Rain and Max Humidity in Minas Gerais')
plt.xlabel('Max Rain (millimeters)')
plt.ylabel('Max Humidity (%)')
plt.grid(True)
plt.show()

#correlation
rain_humidiity = weather[["rain_max", "hum_max"]].corr('pearson')
print(f"Correlation between max rain and max humidity: \n {rain_humidiity}")

#correlation matrix
sns.heatmap(rain_humidiity, annot=True, cmap='viridis').set_title('Correlation between max rain and max humidity')
plt.tight_layout()
plt.show()
```

```python

#Relationship between average wind speed and average temperature

#pivot table 
print("\nPivot table between average wind speed and average temperature:")
pivot = pd.pivot_table(weather, values=['wind_avg','temp_avg'], index = 'year', aggfunc= 'mean')
display(pivot)

#scatterplot
sns.scatterplot(weather, x = "wind_avg", y  = "temp_avg", color = "blue")
plt.title('Relationship between Average Wind Speed and Average Temperature in Minas Gerais')
plt.xlabel('Average Wind Speed (Meters per Second)')
plt.ylabel('Average Temperature (C)')
plt.grid(True)
plt.show()

#correlation
windspeed_temp = weather[["wind_avg", "temp_avg"]].corr()
print(f"Correlation between average wind speed and average temperature: \n {windspeed_temp}")
```
